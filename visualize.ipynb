```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read more normal data
running_data = pd.read_csv('./data/normal/running_data.csv')
sitting_data = pd.read_csv('./data/normal/sitting_data.csv')
walking_data = pd.read_csv('./data/normal/walking_data.csv')
writing_data = pd.read_csv('./data/normal/writing_data.csv')
# Read falling data
falling_data = pd.read_csv('./data/falling/falling_data.csv')
```

```python
# Visualize the comparison of the running data and falling data throught the accelerometer data
plt.figure(figsize=(10, 5))
plt.plot(running_data['AccX'], label='Running')
plt.plot(walking_data['AccX'], label='Walking')
plt.plot(sitting_data['AccX'], label='Sitting')
plt.plot(writing_data['AccX'], label='Writing')
plt.plot(falling_data['AccX'], label='Falling')
plt.title('Comparison of the normal data and falling data throught the accelerometer data')
plt.legend()
```
